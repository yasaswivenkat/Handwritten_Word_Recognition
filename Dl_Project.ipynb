```python
!pip install kagglehub scikit-learn tensorflow

# Step 1: Importing required libraries
import os
from google.colab import userdata
from google.colab import auth
from google.colab import files
from google.colab import drive
import kagglehub

# Step 2: Loading Kaggle credentials
try:
    os.environ["KAGGLE_USERNAME"] = userdata.get('KAGGLE_USERNAME')
    os.environ["KAGGLE_KEY"] = userdata.get('KAGGLE_KEY')
except Exception:
    print("Kaggle credentials not found in Colab secrets!!")

#Step 3: Downloading IAM dataset from kaggle
try:
    print("Downloading and unzipping dataset...")
    path = kagglehub.dataset_download("naderabdalghani/iam-handwritten-forms-dataset")
    !kaggle datasets download -d naderabdalghani/iam-handwritten-forms-dataset -p /content
    !unzip -o /content/iam-handwritten-forms-dataset.zip -d /content/data
    print("Dataset unzipped to /content/data")
except Exception as e:
    print(f"Could not automatically download/unzip: {e}")
    print("Assuming dataset is manually placed,try with adjusting path.")
    path = "/content/data" # Default path
print("Path to dataset files:", path)
try:
    print("Files in dataset root:", os.listdir(path)[:5])
except FileNotFoundError:
    print("Warning: /content/data not found. Please ensure that dataset is unzipped or path is adjusted correctly")

# Main Code :
import os
import glob
import random
import xml.etree.ElementTree as ET
from collections import Counter
import numpy as np
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report

# ---------------------------
ROOT_DIR=os.path.join("/content", "data")
MAX_SAMPLES=10000
IMG_H, IMG_W=64,256
TOP_K_WORDS=20
BATCH_SIZE=32
EPOCHS=100
RANDOM_SEED=42
TRANSFER_MODELS=['VGG16','ResNet50V2','EfficientNetB0'] # 3 pre-trained models

tf.random.set_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)

# ---------------------------
def find_files(root):
    matches = []
    for ext in ["png", "jpg", "tif", "tiff", "jpeg"]:
        matches.extend(glob.glob(os.path.join(root, "**", f"*.{ext}"), recursive=True))
    xmls = glob.glob(os.path.join(root, "**", "*.xml"), recursive=True)
    return matches, xmls

all_images, all_xmls = find_files(ROOT_DIR)
print("-" * 50)
print("Using ROOT_DIR =", ROOT_DIR)
print("Found images:", len(all_images), "Found xmls:", len(all_xmls))

if len(all_images) == 0:
    print("No images found!!")
    all_images, all_xmls = find_files("/content")
    ROOT_DIR = "/content"
    print(f"Found {len(all_images)} images in /content.")
    if len(all_images) == 0:
        raise FileNotFoundError("No images found. Please verify dataset unzipped and path (ROOT_DIR).")

# ---------------------------
# Extract labels
pairs = []
def parse_xml_for_pairs(xml_paths, img_root):
    extracted = []
    for xp in xml_paths:
        try:
            tree = ET.parse(xp)
        except Exception:
            continue
        root = tree.getroot()
        for word_tag in root.iter():
            if word_tag.tag.lower().endswith("word"):
                text = word_tag.attrib.get('text') or word_tag.attrib.get('content') \
                        or word_tag.attrib.get('name') or word_tag.attrib.get('label')
                if text is None:
                    text = (word_tag.text or "").strip()
                file_ref = word_tag.attrib.get('file') or word_tag.attrib.get('image')
                if file_ref:
                    full = os.path.join(img_root, file_ref) if not os.path.isabs(file_ref) else file_ref
                else:
                    full = None
                # Checking for image existence
                if text and full:
                    if not os.path.exists(full):
                         parts = os.path.dirname(xp).split(os.sep)
                         if len(parts) >= 2:
                             sub_dir_guess = os.sep.join(parts[:-2] + ["words", full])
                             if os.path.exists(sub_dir_guess):
                                full = sub_dir_guess
                    if os.path.exists(full):
                        extracted.append((full, text))
    print(f"Extracted {len(extracted)} pairs from XMLs.")
    return extracted

if all_xmls:
    print("Attempting to parse XMLs for labels...")
    xml_pairs = parse_xml_for_pairs(all_xmls, ROOT_DIR)
    print("Pairs recovered from xml:", len(xml_pairs))
    pairs.extend(xml_pairs)
if not pairs:
    print("Falling back to filename heuristics for labels.")
    for p in all_images:
        fname = os.path.basename(p)
        label = None
        for sep in ["-", "_", "."]:
            parts = fname.split(sep)
            if len(parts) >= 2:
                candidate = parts[0]
                if any(c.isalpha() for c in candidate):
                    label = candidate.lower()
                    break
        if label is None:
            label = os.path.splitext(fname)[0].lower()
        pairs.append((p, label))
    print("Pairs extracted:", len(pairs))
random.shuffle(pairs)
pairs = pairs[:MAX_SAMPLES]
print("Total pairs used after initial filtering and trimming:", len(pairs))

# ---------------------------
# Building word frequency and filter
words = [w for (_, w) in pairs]
freq = Counter(words)
top_words = [w for w, _ in freq.most_common(TOP_K_WORDS)]
print(f"Top {TOP_K_WORDS} words: {top_words}")

pairs = [(p, w) for (p, w) in pairs if w in top_words]
print("Pairs after filtering for top K words:", len(pairs))

if len(pairs) == 0:
    raise ValueError("No pairs left after filtering top K words.")

# ---------------------------
# Loading image arrays and Preprocessing
def load_and_preprocess(pairs, img_h=IMG_H, img_w=IMG_W):
    X, Y = [], []
    for p, w in pairs:
        try:
            img = cv2.imread(p, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue
            # Padding
            img = cv2.resize(img, (img_w, img_h))
            img = img.astype(np.float32) / 255.0
            X.append(img)
            Y.append(w)
        except Exception:
            continue
    return np.array(X), np.array(Y)
X, Y = load_and_preprocess(pairs)
print("-" * 50)
print("Loaded images shape:", X.shape)
if len(X) == 0:
    raise ValueError("No images loaded after preprocessing.")
Xc = np.expand_dims(X, -1)
le_words = LabelEncoder()
y_cls = le_words.fit_transform(Y)
num_classes = len(le_words.classes_)
print("Number of classes:", num_classes)
class_names = le_words.classes_
X_train, X_test, y_train, y_test = train_test_split(Xc, y_cls, test_size=0.2, random_state=RANDOM_SEED, stratify=y_cls)
print("Train/Test shapes (Grayscale):", X_train.shape, X_test.shape)

# ---------------------------
#  For Transfer Learning models: gray → 3-channel
def gray_to_3ch(Xgray):
    return np.repeat(Xgray, 3, axis=-1)
X_train_3 = gray_to_3ch(X_train)
X_test_3  = gray_to_3ch(X_test)
print("Train/Test shapes (3-Channel):", X_train_3.shape, X_test_3.shape)
results = {}

# ---------------------------
# Model 1: CNN Classifier
print("\n" + "=" * 50)
print("Training Model 1: CNN Classifier")
print("=" * 50)

def build_cnn_classifier(input_shape, n_classes):
    inp = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, 3, activation='relu', padding='same')(inp)
    x = layers.MaxPool2D(2)(x)
    x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)
    x = layers.MaxPool2D(2)(x)
    x = layers.Conv2D(128, 3, activation='relu', padding='same')(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(128, activation='relu')(x)
    out = layers.Dense(n_classes, activation='softmax')(x)
    m = models.Model(inp, out, name='cnn_classifier')
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return m

cnn_model = build_cnn_classifier((IMG_H, IMG_W, 1), num_classes)
cnn_model.summary()
history_cnn = cnn_model.fit(X_train, y_train, validation_data=(X_test, y_test),
                            epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=2)

# ---------------------------
# CNN → LSTM and GRU models

def cnn_feature_extractor(input_shape):
    inp = layers.Input(shape=input_shape)
    x = layers.Conv2D(32, 3, activation='relu', padding='same')(inp)
    x = layers.MaxPool2D((2, 1))(x)
    x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)
    x = layers.MaxPool2D((2, 1))(x)
    return models.Model(inp, x, name='cnn_feat')

feat_extractor = cnn_feature_extractor((IMG_H, IMG_W, 1))

def X_to_seq_feats(model_feat, X):
    print("Extracting features...")
    feat = model_feat.predict(X, batch_size=BATCH_SIZE, verbose=0)
    n, h, w, c = feat.shape
    return feat.reshape((n, w, h * c))

X_train_seq = X_to_seq_feats(feat_extractor, X_train)
X_test_seq  = X_to_seq_feats(feat_extractor, X_test)
print(f"Sequence feature shape: {X_train_seq.shape}")

print("\n" + "=" * 50)
print("Training Model 2: CNN-LSTM")
print("=" * 50)

def build_lstm_seq_model(seq_shape, n_classes):
    inp = layers.Input(shape=seq_shape)
    x = layers.Bidirectional(layers.LSTM(128))(inp)
    x = layers.Dense(128, activation='relu')(x)
    out = layers.Dense(n_classes, activation='softmax')(x)
    m = models.Model(inp, out, name='cnn_lstm')
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return m

lstm_model = build_lstm_seq_model((X_train_seq.shape[1], X_train_seq.shape[2]), num_classes)
history_lstm = lstm_model.fit(X_train_seq, y_train, validation_data=(X_test_seq, y_test),
                             epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=2)

print("\n" + "=" * 50)
print("Training Model 3: CNN-GRU")
print("=" * 50)

def build_gru_seq_model(seq_shape, n_classes):
    inp = layers.Input(shape=seq_shape)
    x = layers.Bidirectional(layers.GRU(128))(inp)
    x = layers.Dense(128, activation='relu')(x)
    out = layers.Dense(n_classes, activation='softmax')(x)
    m = models.Model(inp, out, name='cnn_gru')
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return m

gru_model = build_gru_seq_model((X_train_seq.shape[1], X_train_seq.shape[2]), num_classes)
history_gru = gru_model.fit(X_train_seq, y_train, validation_data=(X_test_seq, y_test),
                            epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=2)

# ---------------------------
# Models 4,5,6: Transfer Learning

def build_transfer_model(base_model_name, input_shape, n_classes):
    print(f"Building Transfer Model: {base_model_name}")
    inp = layers.Input(shape=input_shape)

    if base_model_name == 'VGG16':
        base = tf.keras.applications.VGG16(include_top=False, weights='imagenet', input_tensor=inp)
        preprocess = tf.keras.applications.vgg16.preprocess_input
    elif base_model_name == 'ResNet50V2':
        base = tf.keras.applications.ResNet50V2(include_top=False, weights='imagenet', input_tensor=inp)
        preprocess = tf.keras.applications.resnet_v2.preprocess_input
    elif base_model_name == 'EfficientNetB0':
        base = tf.keras.applications.EfficientNetB0(include_top=False, weights='imagenet', input_tensor=inp)
        preprocess = tf.keras.applications.efficientnet.preprocess_input
    else:
        raise ValueError(f"Unknown model name: {base_model_name}")
    # Freezing the base model layers
    base.trainable = False

    x = base.output
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    out = layers.Dense(n_classes, activation='softmax')(x)
    m = models.Model(inputs=inp, outputs=out, name=f'{base_model_name}_transfer')
    m.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return m, preprocess

transfer_histories = {}
model_names = []
# Preprocessing 3-channel data
X_train_3_proc = {}
X_test_3_proc = {}
transfer_models = {}
for name in TRANSFER_MODELS:
    print("\n" + "=" * 50)
    print(f"Training Model: {name} Transfer (Model {4 + len(transfer_models)})")
    print("=" * 50)
    try:
        model, preprocess_fn = build_transfer_model(name, (IMG_H, IMG_W, 3), num_classes)
        model_names.append(model.name)
        transfer_models[name] = model
        X_train_3_proc[name] = preprocess_fn(X_train_3 * 255.0)
        X_test_3_proc[name] = preprocess_fn(X_test_3 * 255.0)

        history = model.fit(X_train_3_proc[name], y_train, validation_data=(X_test_3_proc[name], y_test),
                            epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=2)
        transfer_histories[name] = history
    except Exception as e:
        print(f"Skipping {name} due to error: {e}")

# ---------------------------
# Final Results
print("\n" + "=" * 50)
print("FINAL MODEL EVALUATION")
print("=" * 50)
models_list = [('CNN Classifier', cnn_model, X_test, y_test),
              ('CNN-LSTM', lstm_model, X_test_seq, y_test),
              ('CNN-GRU', gru_model, X_test_seq, y_test),]
for name in TRANSFER_MODELS:
    if name in transfer_models:
        models_list.append((f'{name} Transfer', transfer_models[name], X_test_3_proc[name], y_test))

all_metrics = {}
for model_name, model, X_data, y_true in models_list:
    print(f"\n--- Evaluating {model_name} ---")
    # 1. Prediction and Accuracy
    y_pred_probs = model.predict(X_data, verbose=0)
    y_pred_cls = np.argmax(y_pred_probs, axis=1)
    test_accuracy = accuracy_score(y_true, y_pred_cls)
    print(f"Test Accuracy: {test_accuracy:.4f}")

    # 2. Detailed Metrics like Precision,Recall,F1-Score
    report = classification_report(y_true, y_pred_cls, target_names=class_names, zero_division=0, output_dict=True)
    all_metrics[model_name] = {
        'Accuracy': test_accuracy,
        'Precision (Macro)': report['macro avg']['precision'],
        'Recall (Macro)': report['macro avg']['recall'],
        'F1-Score (Macro)': report['macro avg']['f1-score'],
        'Classification Report': classification_report(y_true, y_pred_cls, target_names=class_names, zero_division=0)}

    # 3. Confusion Matrix
    cm = confusion_matrix(y_true, y_pred_cls)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)

    # Plotting the confusion matrix
    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(cmap=plt.cm.Blues, ax=ax)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()

# ---------------------------
# Accuracies Summary
print("\n" + "#" * 50)
print("ACCURACIES AND METRICS SUMMARY")
print("#" * 50)
print(f"Total samples used: {len(y_true) + len(y_train)}")
print(f"Number of classes (Top {TOP_K_WORDS} words): {num_classes}")
print("\n| Model Name | Test Accuracy | Macro Precision | Macro Recall | Macro F1-Score |")
print("|:---|:---:|:---:|:---:|:---:|")
best_accuracy = 0
best_model = ""
for model_name, metrics in all_metrics.items():
    acc = metrics['Accuracy']
    prec = metrics['Precision (Macro)']
    rec = metrics['Recall (Macro)']
    f1 = metrics['F1-Score (Macro)']
    print(f"| {model_name:<12} | {acc:.4f} | {prec:.4f} | {rec:.4f} | {f1:.4f} |")
    if acc > best_accuracy:
        best_accuracy = acc
        best_model = model_name

print("\n" + "-" * 50)
print(f"The best performing model is: {best_model} with Accuracy: {best_accuracy:.4f}")
print("-" * 50)
if best_model in all_metrics:
    print(f"\nDetailed Classification Report for {best_model}:\n")
    print(all_metrics[best_model]['Classification Report'])
```
